```python
# Import Dependencies
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome',**executable_path, headless=False)
```

```python
#Visit the site with our prescription drug dispense rate table  CHANGE the year in the url to get each one... probably could've made this a loop
browser.visit("https://www.cdc.gov/drugoverdose/rxrate-maps/county2020.html")
```

```python
# Create a Beautiful Soup object
html = browser.html
news_soup = soup(html,'html.parser')
```

```python
#all of our data is contained inside the tbody tag and then each row within.  Some rows have clases some don't

our_body = news_soup.find_all("tbody")
our_rows = our_body[0].find_all("tr")
#len(our_rows)
#our_rows[0]
```

```python
#Build an array of arrays to contain the data

our_data = []
for row in our_rows:
    data_points = row.find_all("td")
    row_data = [data.text for data in data_points]
    our_data.append(row_data)

our_data
```

```python
#adding our year, make sure it matches the year from the url and output
for row in our_data:
    row.append(2020)
```

```python
#turn our data into a dataframe
#this is for 2017, 2019, and 2021
tempdata_df = pd.DataFrame(data=our_data, columns=['State','County','County_FIPSCode','OpioidDispenseRateper100',"Year"])

#this is for 2018, for some reason it has county then state, but we will reorder it to match all the rest for later
#tempdata_df = pd.DataFrame(data=our_data, columns=['County','State','County_FIPSCode','OpioidDispenseRateper100',"Year"])
#tempdata_df = tempdata_df[['State','County','County_FIPSCode','OpioidDispenseRateper100',"Year"]]


tempdata_df.head()
```

```python
#Export the data, don't forget to have the title in our output here match which page we are on
tempdata_df.to_csv("Resources/2020CountyOpioidPrescriptionRate.csv",index=False)
```
