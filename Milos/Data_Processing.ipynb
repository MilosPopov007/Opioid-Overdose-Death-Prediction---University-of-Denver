```python
# Add a year column to each data frame
opioid_deaths_2016["Year"] = 2016
opioid_deaths_2017["Year"] = 2017
opioid_deaths_2018["Year"] = 2018
opioid_deaths_2019["Year"] = 2019
opioid_deaths_2020["Year"] = 2020

# Merge the data frames into one
opioid_deaths_all = pd.concat([opioid_deaths_2016, opioid_deaths_2017, opioid_deaths_2018, opioid_deaths_2019, opioid_deaths_2020], ignore_index=True)

# Print the first few rows of the merged data frame
opioid_deaths_all.tail()
```

```python
# Drop last row containing summary statistics
opioid_deaths_all = opioid_deaths_all.drop(opioid_deaths_all.tail(1).index)

# Display tail of merged dataframe to verify that last row was dropped
opioid_deaths_all.tail()
```

```python
# Export

opioid_deaths_all.to_csv('opioid_deaths_all.csv', index=False)
```

```python
# Replace "Suppressed" values with NaN
opioid_deaths_all['Deaths'] = pd.to_numeric(opioid_deaths_all['Deaths'], errors='coerce')
opioid_deaths_all['Crude Rate'] = pd.to_numeric(opioid_deaths_all['Crude Rate'], errors='coerce')

# Check for any remaining non-numeric values
print(opioid_deaths_all.isna().sum())

# Drop any rows with missing values
opioid_deaths_all.dropna(inplace=True)

# Check the cleaned dataframe
opioid_deaths_all.tail()
```

```python
# Reorder the columns
opioid_deaths_all = opioid_deaths_all[['Year', 'County', 'County Code', 'Deaths', 'Population', 'Crude Rate']]

# Reset the index and drop the old index
opioid_deaths_all = opioid_deaths_all.reset_index(drop=True)
opioid_deaths_all.head()
```

```python
# Create a new dataframe called opioid_deaths_clean by dropping all rows with missing values in any column
# create a copy of the original dataframe
opioid_deaths_clean1 = opioid_deaths_all.copy()

# drop rows with missing values
opioid_deaths_clean1.dropna(inplace=True)

# display the cleaned dataframe
opioid_deaths_clean1.head()

# Option 1: Dropping Rows with Missing Values

# Explanation:

# This code drops all rows with missing values in any column of the opioid_deaths_all dataframe
# and assigns the resulting cleaned dataframe to a new variable called opioid_deaths_clean.
# This approach is useful when there are relatively few missing values,
# and it is feasible to remove them without sacrificing too much of the data.
# One potential drawback is that it can result in a smaller dataset,
# which may negatively impact the performance of a machine learning model that is trained on the data.
```

```python
# create a copy of the original dataframe
opioid_deaths_clean2 = opioid_deaths_all.copy()

# convert 'Deaths' and 'Crude Rate' columns to numeric data types
opioid_deaths_clean2['Deaths'] = pd.to_numeric(opioid_deaths_clean2['Deaths'], errors='coerce')
opioid_deaths_clean2['Crude Rate'] = pd.to_numeric(opioid_deaths_clean2['Crude Rate'], errors='coerce')

# impute missing values with the median of each column
opioid_deaths_clean2['Deaths'].fillna(opioid_deaths_clean2['Deaths'].median(), inplace=True)
opioid_deaths_clean2['Crude Rate'].fillna(opioid_deaths_clean2['Crude Rate'].median(), inplace=True)

# display the cleaned dataframe
opioid_deaths_clean2.head()

# Option 2: Imputing Missing Values

# This code creates a copy of the original dataframe to avoid modifying the original data.
# It then uses pd.to_numeric() to convert the 'Deaths' and 'Crude Rate' columns to numeric data types,
# with any non-numeric values replaced with NaN.
# Finally, it uses the fillna() method to impute the missing values with the median of each column.
# This approach is useful when there are many missing values, and it is necessary to impute them to avoid losing too much data.
# Potential drawback is that it can introduce bias into the data,
# particularly if the missing values are not missing at random.
```
